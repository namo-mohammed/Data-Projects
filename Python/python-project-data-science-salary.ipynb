## Charts

```python
# Create a Figure with 2 Subplots (1 Row, 2 Columns)
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# First Subplot: Bar Chart for Work Year
counts_work_year = data["Work Year"].value_counts().sort_index()
bars1 = ax[0].bar(counts_work_year.index, counts_work_year.values, color=["#FF5733", "#33FF57", "#3357FF", "#F3FF33"], edgecolor="black", linewidth=1.2)
ax[0].set_title("Job Count for Each Work Year", fontsize=12, fontweight="bold")
ax[0].set_xlabel("Work Year", fontsize=12)
ax[0].set_ylabel("Amount of Jobs", fontsize=12)
ax[0].set_xticks(counts_work_year.index)
ax[0].set_xticklabels(counts_work_year.index, rotation=45)

# Adding Counts on Top of the Bars for the First Subplot
ax[0].bar_label(bars1, label_type='edge', fontsize=10, fontweight='bold')

# Second Subplot: Bar chart for Company Size
counts_company_size = data["Company Size"].value_counts().sort_index()
bars2 = ax[1].bar(counts_company_size.index, counts_company_size.values, color=["#FF5733", "#33FF57", "#3357FF", "#F3FF33"], edgecolor="black", linewidth=1.2)
ax[1].set_title("Job Count for Company Sizes", fontsize=12, fontweight="bold")
ax[1].set_xlabel("Company Size", fontsize=12)
ax[1].set_ylabel("Amount of Jobs", fontsize=12)
ax[1].set_xticks(range(len(counts_company_size)))
ax[1].set_xticklabels(counts_company_size.index, rotation=45)

# Adding Counts on Top of the Bars for the Second Subplot
ax[1].bar_label(bars2, label_type='edge', fontsize=10, fontweight='bold')

# Adjust Layout to Prevent Overlap
plt.tight_layout()

# Show the Plots
plt.show()
```

**Insights**

We find data for the year 2020 even though the description of the dataset aims to shed light on the salary trends in the field of Data Science for the years 2021 to 2023.

The year 2023 had the highest amount of Data Science jobs. In total there were 1789 jobs available worldwide. We can see an increase in jobs available each year. 2020 we only have 76 jobs and then we leap forward to 2022 & 2023 with a significant difference with figures in the thousands.

In addition to this, Medium Size companies had the greatest number of jobs available with a total of 3157 jobs across the board.

```python
# Get the Top 20 Salaries
top_20_salaries = data.nlargest(20, 'Salary in USD')

# Create a Unique Identifier for Plotting Purposes
top_20_salaries['Unique Job Title'] = top_20_salaries['Job Title'] + ' (' + top_20_salaries['Company Location'] + ') '
top_20_salaries['Unique Job Title'] += top_20_salaries.index.astype(str)

# Get the Bottom 20 Salaries
bottom_20_salaries = data.nsmallest(20, 'Salary in USD')

# Create a Unique Identifier for Plotting Purposes, including Employment Type
bottom_20_salaries['Unique Job Title'] = (bottom_20_salaries['Job Title'] + ' (' +
                                          bottom_20_salaries['Company Location'] + ', ' +
                                          bottom_20_salaries['Employment Type'] + ') ' +
                                          bottom_20_salaries.index.astype(str))

# Create Subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12,14))

# Plotting the Top 20 Salaries Bar Chart with Customizations
bars_top = axes[0].barh(top_20_salaries['Unique Job Title'], top_20_salaries['Salary in USD'], color='mediumseagreen')

# Remove the Lines at the end of the Bars by setting Edgecolor to 'none' and Linewidth to 0
for bar in bars_top:
    bar.set_edgecolor('none')
    bar.set_linewidth(0)

# Add Borders to the Columns
for bar in bars_top:
    bar.set_edgecolor('black')
    bar.set_linewidth(1)
    
# Use the Original Job Titles with Company Location for the y-axis Labels
axes[0].set_yticks(range(len(top_20_salaries)))
axes[0].set_yticklabels(top_20_salaries['Job Title'] + ' (' + top_20_salaries['Company Location'] + ')')

axes[0].set_xlabel('Salary in $ USD', fontsize=12)
axes[0].set_ylabel('Job Title (Company Location)', fontsize=12)
axes[0].set_title('Top 20 Data Science Salaries by Job Title and Location', fontsize=14, fontweight='bold')
axes[0].invert_yaxis() # Invert y-axis to have the Highest Salary on Top

# Plotting the Bottom 20 Salaries Bar Chart with Customizations
bars_bottom = axes[1].barh(bottom_20_salaries['Unique Job Title'], bottom_20_salaries['Salary in USD'], color='darkorange')

# Remove the Lines at the end of the Bars by setting Edgecolor to 'none' and Linewidth to 0
for bar in bars_bottom:
    bar.set_edgecolor('none')
    bar.set_linewidth(0)

# Add Borders to the Columns
for bar in bars_bottom:
    bar.set_edgecolor('black')
    bar.set_linewidth(1)
    
# Use the Original Job Titles with Company Location and Employment Type for the y-axis Labels
axes[1].set_yticks(range(len(bottom_20_salaries)))
axes[1].set_yticklabels(bottom_20_salaries['Job Title'] + ' (' +
                        bottom_20_salaries['Company Location'] + ', ' +
                        bottom_20_salaries['Employment Type'] + ')')

axes[1].set_xlabel('Salary in $ USD', fontsize=12)
axes[1].set_ylabel('Job Title (Company Location, Employment Type)', fontsize=12)
axes[1].set_title('Bottom 20 Data Science Salaries by Job Title, Location and Employment Type', fontsize=14, fontweight='bold')
axes[1].invert_yaxis() # Invert y-axis to have the highest salary on top

# Adjust Layout and Show the Plot
plt.tight_layout()
plt.show()
```

**Insights**

The United States dominates the list of the highest-paying Data Science roles, with 18 out of the top 20 positions based there, while the remaining two are in Great Britain and Israel. Salaries for these roles range from 300,000 to over 400,000 USD.

On the lower end, the least-paying Data Science jobs are primarily found in Asia, Africa, and South America, offering salaries between 4,000 and 8,000 USD.

```python
# Calculate the Counts of Each Experience Level
work_year_counts = data['Experience Level'].value_counts()

# Ensure the Explode Parameter Matches the Number of Slices
explode = (0.1, 0, 0) if len(work_year_counts) == 3 else (0.1, 0, 0, 0)

# Define a Color Palette
palette = sns.color_palette("pastel", len(work_year_counts))

# Function to Display Both Percentage and Count
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:d})'
    return my_format

# Plotting the Pie Chart with Additional Customizations
plt.figure(figsize=(7,7))
plt.pie(work_year_counts, 
        labels=work_year_counts.index, 
        autopct=autopct_format(work_year_counts), 
        colors=palette, 
        explode=explode, 
        shadow=True,
        textprops={'fontweight': 'light'})  # Make labels bold

plt.title('Experience Level Distribution For All Data Science Jobs', fontsize=14, fontweight='bold')

# Add a Legend
plt.legend(title="Experience Level", loc="upper right")

# Display the Plot
plt.show()
```
